```python
from keras.models import load_model
import joblib
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
import pandas as pd


model = load_model("model.h5")
tokenizer = joblib.load("tokenizer.pkl")


def predictive_system(review):
  sequences=tokenizer.texts_to_sequences([review])
  padded_sequences = pad_sequences(sequences,maxlen=200)
  prediction=model.predict(padded_sequences)
  sentiment = "positive" if prediction [0][0] > 0.5 else "negative"
  return sentiment
```

```python
review_sentiment = predictive_system("the movie was very good at first half whole actors mainly main charcter was too good and dances and song and graphics and story but when it comes to 2nd half the movie is lag and predictable secnes ")
review_sentiment
```

```python
import gradio as gr
title  = "Movie Sentiment analysis application"
app = gr.Interface(fn= predictive_system,inputs="textbox",outputs="text",title=title)
app.launch(share = True)
```
